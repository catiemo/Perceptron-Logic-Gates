### Importing Libraries

```python
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
import numpy as np
from itertools import product
```

### Creating Data and Labels

```python
data = [[0, 0], [0, 1], [1, 0], [1, 1]]
labels = [0, 0, 0, 1]
```

### Visualizing Data points

```python
plt.scatter([point[0] for point in data], [point[1] for point in data], c = labels)
plt.show()
```

### Building the Perceptron

```python
classifier = Perceptron(max_iter = 40)
```

### Train the Model

```python
classifier.fit(data, labels)
```

### Testing the Model

```python
classifier.score(data, labels)
```

### Distance from the Decision Boundary

```python
classifier.decision_function([[0, 0], [1, 1], [0.5, 0.5]])
```

### Visualizing the Perceptron

```python
x_values = np.linspace(0, 1, 100)
y_values = np.linspace(0, 1, 100)
point_grid = list(product(x_values, y_values))
point_grid
```

```python
distances = classifier.decision_function(point_grid)
distances
```

```python
abs_distances = [abs(distance) for distance in distances]
distances_matrix = np.reshape(abs_distances, (100, 100))
```

```python
heatmap = plt.pcolormesh(x_values, y_values, distances_matrix, shading = 'auto')
plt.colorbar(heatmap)
plt.show()
```
